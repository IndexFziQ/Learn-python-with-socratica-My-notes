布尔变量一般用于比较，举个例子：

```
a = 4
b = 5
a == b
```

上面这个例子中，我们用一个等号来赋值，用两个等号来比较二值是否相等，其返回值就是布尔变量。如果想要返回True也很简单，这些只要在python数值运算的部分找表格。



```
a!=b
```
