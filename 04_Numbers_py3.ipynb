## float类型

```
# define float number e
e = 2.71821291

# To confirm it is a float, get its type.
type(e)
```

## complex类型

```
# define complex number z
z = 6 - 1.4j

# check its type
type(z)
```

```
# get its's real part and imag part
print(z.real)
print(z.imag)
```

和python2一样，都是用float类型来表示实部和虚部的。
